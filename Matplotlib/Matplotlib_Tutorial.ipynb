### Pyplot API and Object-Oriented (OO) API
* Pyplot API -> Quickly
* OO API -> Advanced

#### `Pyplot API`

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)")
plt.plot(x, np.cos(x), color="red", linestyle="solid", label="cos(x)")
plt.title("sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
# plt.xlim([0, 4])
# plt.ylim([-0.75, 0.75])
plt.axis([0, 4, -1, 1]) # <=> plt.xlim([0, 4]) plt.ylim([-0.75, 0.75]) (plt.axis([xmin, xmax, ymin, ymax]))
# plt.axis("tight") 
# plt.axis("equal")
plt.legend()
plt.show()
```

#### `Object-Oriented API`

```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5));

# 3. plot the data
ax.plot(x, y)
ax.set_title("simple plot")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

plt.show();
```

### Most common types of Matplotlib plots
- line
- scatter
- bar
- hist
- subplots()

#### `Line`

```python
x = np.linspace(0, 10, 100)
x[:5]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

ax.plot(x, x**3);
```

#### `Scatter`

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

sizes = 1000 * rng.randn(100)
colors = rng.randn(100)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
img = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.5)
fig.colorbar(img)
plt.show();
```

#### `Bar`
- Vertical
- Horizontal

```python
soft_drink_prices = {
    "Coke": 10,
    "Pessi": 12,
    "Sprite": 8
}

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color="green")
ax.set_title("Bach Hoa Xanh's Soft Drink Prices")
ax.set_ylabel("Prices = ($)")
plt.show();
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
plt.show()
```

#### `Histogram`

```python
student_height = np.random.normal(170, 10, size=250)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

ax.hist(student_height, bins =30, color="green", edgecolor="black");
```

#### `Pie`

```python
y = np.array([40, 20, 25, 15])
mylabels = ["Orange", "Banana", "Kiwi", "Apple"]
myexplode = [0, 0, 0, 0.1]
plt.pie(y, labels=mylabels, explode=myexplode, shadow=True, autopct="%1.1f%%")
plt.title("This is a fruit distribution")
plt.legend(title="Fruit Type", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # bbox_to_anchor=(x0, y0, width, height) (0: left, 1: right)
plt.show()
```

### Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

for ax in axes.flat:
    ax.grid(True, axis="y", linestyle="--", alpha=0.5)

axes[0, 0].plot(x, x/2, c="green")
axes[0, 0].plot(x, -x*2, c="red")

axes[0, 1].scatter(np.random.random(10), np.random.random(10))

axes[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values())

axes[1, 1].hist(student_height, bins=30)

plt.show()
```
