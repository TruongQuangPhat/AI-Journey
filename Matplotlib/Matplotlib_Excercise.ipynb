```python
time = np.array(["6h", "9h", "12h", "15h", "18h"])
temperature = np.array([22, 24, 30, 29, 22])

plt.plot(time, temperature, marker="o", linestyle="-", color="green", linewidth="2")
plt.title("Temperature & Time")
plt.xlabel("Time")
plt.ylabel("Temperatute (C)")
plt.grid(True, axis="y")
plt.show()
```

4. Vẽ nhiều đường trên cùng một biểu đồ
Yêu cầu: Thể hiện nhiệt độ của 2 thành phố trên cùng một biểu đồ.

```python
time = np.array([0, 3, 6, 9, 12, 15, 18, 21])
temperature_HCM = np.array([18, 19, 22, 25, 32, 30, 28, 26])
temperature_HN = np.array([15, 18, 22, 24, 26, 25, 21, 20])
yticks = np.arange(15, 37, 2)
plt.figure(figsize=(8, 5))
plt.style.use("seaborn-v0_8-darkgrid")
plt.plot(time, temperature_HCM, marker="*", linestyle="--", label="Temperature_HCM", color="green")
plt.plot(time, temperature_HN, marker="x", linestyle="-.", label="Temperature_HN", color="skyblue")
plt.xticks(time, [f"{t}h" for t in time], rotation=-30) # ticks, labels, 
plt.yticks(yticks, [f"{y}°C" for y in yticks], rotation=30)
plt.xlabel("Time")
plt.ylabel("Temperrature")
plt.title("Temperature HCM & HN city")
plt.grid(True, axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.axis("tight")
plt.legend(title="City", loc="best", frameon=True) # loc="best" tu dong set 1 vi tri phu hop, frameone hien thi box bao quanh legend
plt.show()
```

5. Biểu đồ cột đơn giản
Yêu cầu: Vẽ biểu đồ cột thể hiện điểm của học sinh trong 5 môn học.

```python
score = {
    "Toán": 10.0,
    "Lý": 9.5,
    "Hóa": 9.0,
    "Sinh": 7.0,
    "Tin": 9.3
}

score_avg = {
    "Toán": 8.6,
    "Lý": 8.2,
    "Hóa": 9.0,
    "Sinh": 8.4,
    "Tin": 8.7
}

plt.figure(figsize=(8, 5))
bars = plt.bar(score.keys(), score.values(), width=0.7, color="skyblue", edgecolor="black", linewidth=1.5, label="Điểm cá nhân")
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, height + 0.1, # vi tri
        f"{height:.1f}", ha="center", va="bottom", fontsize=9, color="black" # ha: cang ngang, va: can doc
    )
plt.plot(score_avg.keys(), score_avg.values(), color="green", linewidth=1.5, label="Điểm trung bình", marker="o")
for x, y in zip(score_avg.keys(), score_avg.values()):
    plt.text(
        x, y + 0.1,
        f"{y:.1f}", ha="center", va="bottom", fontsize=9, color="black"
    )
plt.xlabel("Môn học")
plt.ylabel("Điểm số")
plt.title("Bảng điểm của học sinh Nguyễn Văn A")
#plt.axis("tight")
plt.grid(True, which="both", axis="y", color="gray", linestyle=":", linewidth=0.8)
plt.tight_layout(pad=2.0, w_pad=0.5, h_pad=1.0, rect=[0, 0, 0.9, 1]) # thu nho chart [left, bottom, right, top]  scale
plt.legend(
    loc="upper left", bbox_to_anchor=(1,1),
    ncol=1, fontsize=10, title="Cities", title_fontsize=12,
    frameon=True, facecolor="white", edgecolor="black", framealpha=0.8,
    handlelength=2.5, labelspacing=0.8, columnspacing=1.2, markerscale=1.5
) # bbox_to_anchor = (x, y): neo legend ngoai truc
plt.show()
```

6. Biểu đồ cột nhóm
Yêu cầu: Vẽ biểu đồ cột nhóm so sánh điểm giữa 2 lớp học theo từng môn.

```python
class_A = {
    "Toán": 9.2,
    "Lý": 9.0,
    "Hóa": 8.6
}

class_B = {
    "Toán": 9.0,
    "Lý": 8.5,
    "Hóa": 9.4
}

subjects = list(class_A.keys())
x = np.arange(len(subjects))

width = 0.35

plt.figure(figsize=(8, 5))

bars1 = plt.bar(x - width/2, class_A.values(), width, label="Điểm class A", color="skyblue", edgecolor="black")
bars2 = plt.bar(x + width/2, class_B.values(), width, label="Điểm class B", color="lightgreen", edgecolor="black")

for bar in bars1 + bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f"{height:.1f}", ha="center", va="bottom", fontsize=9)
plt.xticks(x, subjects)
plt.ylim([0, 10])
plt.xlabel("Môn học", fontweight="bold")
plt.ylabel("Điểm số", fontweight="bold")
plt.title("So sánh điểm class A và class B", fontweight="bold")
#plt.axis("tight")
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), frameon=True)
plt.grid(True, linestyle="--", alpha=0.7, axis="y")
plt.show()
```

7. Dùng style có sẵn
Yêu cầu: Sử dụng các style matplotlib có sẵn (ggplot, seaborn...).

```python
"""
Liet ke Style
"""
plt.style.available
```

```python
style_list = plt.style.available
nrows, ncols = 4, 7
X = np.linspace(0, 10, 100)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, 10))

for i, style in enumerate(style_list[:nrows*ncols]):
    ax = axes[i // ncols, i % ncols]
    with plt.style.context(style):
        ax.plot(X, np.sin(X))

        ax.set_title(style, fontweight="bold", fontsize=12)
        
        ax.set_facecolor(plt.rcParams['axes.facecolor'])
        fig.patch.set_facecolor(plt.rcParams['figure.facecolor'])

plt.tight_layout()
plt.show()
```

8. Biểu đồ radar (spider chart)
Yêu cầu: So sánh các kỹ năng của một lập trình viên bằng biểu đồ radar.

```python
import math

skills = ["Python", "C/C++", "SQL", "Algorithms", "System Design", "Testing", "Communication"]
scores = [9, 7, 8, 8, 6, 8, 7]

df = pd.DataFrame({"Skill": skills, "Score": scores})

N = len(skills)

angles = np.linspace(0, 2 * math.pi, N, endpoint=False).tolist()

angles += angles[:1]
values = scores + scores[:1]

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, polar=True)

ax.set_theta_offset(math.pi / 2)
ax.set_theta_direction(-1)

ax.set_thetagrids(np.degrees(angles[:-1]), labels=skills)

ax.set_ylim(0, 10)
ax.set_rlabel_position(180 / N)
ax.set_title("Developer Skills Radar Chart", y=1.08, fontsize=14)

ax.plot(angles, values, linewidth=2)
ax.fill(angles, values, alpha=0.5)

plt.tight_layout()
plt.show()
```
