1. Vẽ biểu đồ đường cơ bản

```python
time = np.array(["6h", "9h", "12h", "15h", "18h"])
temperature = np.array([22, 24, 30, 29, 22])

plt.plot(time, temperature, marker="o", linestyle="-", color="green", linewidth="2")
plt.title("Temperature & Time")
plt.xlabel("Time")
plt.ylabel("Temperatute (C)")
plt.grid(True, axis="y")
plt.show()
```

4. Vẽ nhiều đường trên cùng một biểu đồ
Yêu cầu: Thể hiện nhiệt độ của 2 thành phố trên cùng một biểu đồ.

```python
time = np.array([0, 3, 6, 9, 12, 15, 18, 21])
temperature_HCM = np.array([18, 19, 22, 25, 32, 30, 28, 26])
temperature_HN = np.array([15, 18, 22, 24, 26, 25, 21, 20])
yticks = np.arange(15, 37, 2)
plt.figure(figsize=(8, 5))
plt.style.use("seaborn-v0_8-darkgrid")
plt.plot(time, temperature_HCM, marker="*", linestyle="--", label="Temperature_HCM", color="green")
plt.plot(time, temperature_HN, marker="x", linestyle="-.", label="Temperature_HN", color="skyblue")
plt.xticks(time, [f"{t}h" for t in time], rotation=-30) # ticks, labels, 
plt.yticks(yticks, [f"{y}°C" for y in yticks], rotation=30)
plt.xlabel("Time")
plt.ylabel("Temperrature")
plt.title("Temperature HCM & HN city")
plt.grid(True, axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.axis("tight")
plt.legend(title="City", loc="best", frameon=True) # loc="best" tu dong set 1 vi tri phu hop, frameone hien thi box bao quanh legend
plt.show()
```

5. Biểu đồ cột đơn giản
Yêu cầu: Vẽ biểu đồ cột thể hiện điểm của học sinh trong 5 môn học.

```python
score = {
    "Toán": 10.0,
    "Lý": 9.5,
    "Hóa": 9.0,
    "Sinh": 7.0,
    "Tin": 9.3
}

score_avg = {
    "Toán": 8.6,
    "Lý": 8.2,
    "Hóa": 9.0,
    "Sinh": 8.4,
    "Tin": 8.7
}

plt.figure(figsize=(8, 5))
bars = plt.bar(score.keys(), score.values(), width=0.7, color="skyblue", edgecolor="black", linewidth=1.5, label="Điểm cá nhân")
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, height + 0.1, # vi tri
        f"{height:.1f}", ha="center", va="bottom", fontsize=9, color="black" # ha: cang ngang, va: can doc
    )
plt.plot(score_avg.keys(), score_avg.values(), color="green", linewidth=1.5, label="Điểm trung bình", marker="o")
for x, y in zip(score_avg.keys(), score_avg.values()):
    plt.text(
        x, y + 0.1,
        f"{y:.1f}", ha="center", va="bottom", fontsize=9, color="black"
    )
plt.xlabel("Môn học")
plt.ylabel("Điểm số")
plt.title("Bảng điểm của học sinh Nguyễn Văn A")
#plt.axis("tight")
plt.grid(True, which="both", axis="y", color="gray", linestyle=":", linewidth=0.8)
plt.tight_layout(pad=2.0, w_pad=0.5, h_pad=1.0, rect=[0, 0, 0.9, 1]) # thu nho chart [left, bottom, right, top]  scale
plt.legend(
    loc="upper left", bbox_to_anchor=(1,1),
    ncol=1, fontsize=10, title="Cities", title_fontsize=12,
    frameon=True, facecolor="white", edgecolor="black", framealpha=0.8,
    handlelength=2.5, labelspacing=0.8, columnspacing=1.2, markerscale=1.5
) # bbox_to_anchor = (x, y): neo legend ngoai truc
plt.show()
```

6. Biểu đồ cột nhóm
Yêu cầu: Vẽ biểu đồ cột nhóm so sánh điểm giữa 2 lớp học theo từng môn.

```python
class_A = {
    "Toán": 9.2,
    "Lý": 9.0,
    "Hóa": 8.6
}

class_B = {
    "Toán": 9.0,
    "Lý": 8.5,
    "Hóa": 9.4
}

subjects = list(class_A.keys())
x = np.arange(len(subjects))

width = 0.35

plt.figure(figsize=(8, 5))

bars1 = plt.bar(x - width/2, class_A.values(), width, label="Điểm class A", color="skyblue", edgecolor="black")
bars2 = plt.bar(x + width/2, class_B.values(), width, label="Điểm class B", color="lightgreen", edgecolor="black")

for bar in bars1 + bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f"{height:.1f}", ha="center", va="bottom", fontsize=9)
plt.xticks(x, subjects)
plt.ylim([0, 10])
plt.xlabel("Môn học", fontweight="bold")
plt.ylabel("Điểm số", fontweight="bold")
plt.title("So sánh điểm class A và class B", fontweight="bold")
#plt.axis("tight")
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), frameon=True)
plt.grid(True, linestyle="--", alpha=0.7, axis="y")
plt.show()
```

7. Biểu đồ cột chồng
Yêu cầu: Thể hiện doanh thu từ 2 loại sản phẩm bằng biểu đồ cột chồng.

```python
thang = ["Th1", "Th2", "Th3", "Th4", "Th5"]
doanh_thu_A = [30, 40, 25, 35, 45]
doanh_thu_B = [20, 25, 30, 15, 20]

plt.figure(figsize=(8, 5))


bar1 = plt.bar(thang, doanh_thu_A, label="Sản phẩm A", color="skyblue", edgecolor="black", linewidth=1.5)
bar2 = plt.bar(thang, doanh_thu_B, bottom=doanh_thu_A, label="Sản phẩm B", color="lightgreen", edgecolor="black", linewidth=1.5)

for ba1, ba2 in zip(bar1, bar2):
    plt.text(ba1.get_x() + ba1.get_width()/2, ba1.get_height()/2, f"{ba1.get_height()}", ha="center", va="center", fontsize=9, fontweight="bold")
    plt.text(ba2.get_x() + ba2.get_width()/2, ba1.get_height() + ba2.get_height()/2, f"{ba2.get_height()}", ha="center", va="center", fontsize=9, fontweight="bold")
    

plt.ylabel("Doanh thu (triệu VND)")
plt.title("Biểu đồ cột chồng - Doanh thu theo tháng")
plt.axis("tight")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), frameon=True)
plt.tight_layout()
plt.grid(True, axis="y", linestyle="--", alpha=0.7)
plt.show()
```

8. Biểu đồ tròn
Yêu cầu: Vẽ biểu đồ tròn thể hiện tỷ lệ sử dụng mạng xã hội.

```python
ratio = [40, 20, 15, 15, 10]
labels = ["Tiktok", "Facebook", "Instagram", "Outlook", "Youtube"]
explode = [0, 0, 0, 0, 0.1]
plt.pie(ratio, labels=labels, explode=explode, shadow=True, autopct="%1.1f%%")
plt.legend(title="Social Network", loc="best", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Social media statistics")
plt.tight_layout()
plt.show()
```

9. Biểu đồ histogram
Yêu cầu: Phân phối chiều cao của 100 học sinh bằng biểu đồ histogram.

```python
height_students = np.random.normal(loc=170, scale=15, size=100)

counts, bins, patches = plt.hist(height_students, bins=10, color="deepskyblue", edgecolor="black", linewidth=2)
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Số lượng")
plt.title("Thống kê chiều cao 100 học sinh")
plt.grid(True, axis="y", linestyle="--", alpha=0.7)

count = len(height_students)
min_val = np.min(height_students)
max_val = np.max(height_students)
mean_val = np.mean(height_students)
median_val = np.median(height_students)
std_val = np.std(height_students)

stats_data = [
    ["Số lượng", f"{count}"],
    ["Min", f"{min_val:.2f}"],
    ["Max", f"{max_val:.2f}"],
    ["Trung bình", f"{mean_val:.2f}"],
    ["Trung vị", f"{median_val:.2f}"],
    ["Độ lệch chuẩn", f"{std_val:.2f}"]
]

plt.table(
    cellText=stats_data,
    colLabels=["Thống kê", "Giá trị"],
    loc="upper right",
    cellLoc="center",
    colWidths=[0.3, 0.2],
    bbox=[1.05, 0.2, 0.35, 0.5]
)

plt.tight_layout()
plt.show()
```

10. Biểu đồ phân tán (scatter plot)
Yêu cầu: So sánh chiều cao và cân nặng của 50 người bằng biểu đồ phân tán.

```python
np.random.seed(42)
height = np.random.normal(loc=170, scale=10, size=50)
weight = 0.5 * height + np.random.normal(loc=20, scale=5, size=50)
age = np.random.randint(18, 65, size=50)

plt.scatter(height, weight, c=age, cmap="viridis", edgecolor="black")
plt.colorbar(label="Tuổi")
plt.xlabel("Chiều cao")
plt.ylabel("Cân nặng")
plt.title("Cân nặng & chiều cao (màu thể hiện tuôi)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
